```python
import pandas as pd
df=pd.read_csv("zmeenkarachi.csv")
df
```

```python
df=df[['property_type','price','baths','bedrooms','Area Type','Area Size']]
df
```

```python
area_type_mapping = {'Kanal': 1, 'Marla': 2}
df['Area Type'] = df['Area Type'].map(area_type_mapping)

df
```

```python
area_type_mapping = {'House': 1, 'Flat': 2,'Lower Portion':3,'Upper Portion':4,'room':5,'Penthouse':6}
df['property_type'] = df['property_type'].map(area_type_mapping)

df
```

```python
df['price'].describe()
```

```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt



# Calculating mean and standard deviation
mean_new_data = np.mean(df['price'])
std_dev_new_data = np.std(df['price'])

# Plotting histogram with standard deviation visualization for the new dataset
plt.figure(figsize=(10, 6))
plt.hist(df['price'], bins=15, alpha=0.6, color='b', edgecolor='black')

# Plotting mean and standard deviation lines
plt.axvline(mean_new_data, color='r', linestyle='dashed', linewidth=2, label=f'Mean: {mean_new_data:.2f}')
plt.axvline(mean_new_data - std_dev_new_data, color='g', linestyle='dashed', linewidth=2, label=f'-1 SD: {mean_new_data - std_dev_new_data:.2f}')
plt.axvline(mean_new_data + std_dev_new_data, color='g', linestyle='dashed', linewidth=2, label=f'+1 SD: {mean_new_data + std_dev_new_data:.2f}')

plt.title('Histogram with Standard Deviation (New Dataset)')
plt.xlabel('Data values')
plt.ylabel('Frequency')
plt.legend()
plt.grid(True)

# Show the plot
plt.show()

# Print calculated values
print(f"Mean: {mean_new_data}")
print(f"Standard Deviation: {std_dev_new_data}")
```

```python
# Import necessary libraries
import numpy as np


# Calculate the IQR to identify and remove outliers
Q1 = np.percentile(df["price"], 25)
Q3 = np.percentile(df["price"], 75)
IQR = Q3 - Q1

# Define lower and upper bounds for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Filter data to remove outliers
filtered_data = df["price"][(df["price"] >= lower_bound) & (df["price"] <= upper_bound)]

# Boxplot for filtered data
plt.figure(figsize=(10, 6))
plt.boxplot(filtered_data, patch_artist=True, boxprops=dict(facecolor='lightblue'), showmeans=True)

plt.title('Boxplot After Removing Outliers')
plt.ylabel('Data values')
plt.grid(True)

# Show the plot
plt.show()


```

```python
import numpy as np
from scipy.stats import powerlaw
import matplotlib.pyplot as plt
fig, ax = plt.subplots(1, 1)
```

```python
a = 0.659
mean, var, skew, kurt = powerlaw.stats(a, moments='mvsk')
```

```python
x = np.linspace(powerlaw.ppf(0.01, a),
                powerlaw.ppf(0.99, a), 100)
ax.plot(x, powerlaw.pdf(x, a),
       'r-', lw=5, alpha=0.6, label='powerlaw pdf')
```

```python
rv = powerlaw(a)
ax.plot(x, rv.pdf(x), 'k-', lw=2, label='frozen pdf')
```

```python
vals = powerlaw.ppf([0.001, 0.5, 0.999], a)
np.allclose([0.001, 0.5, 0.999], powerlaw.cdf(vals, a))
```

```python
r = powerlaw.rvs(a, size=1000)
```

```python
ax.hist(r, density=True, bins='auto', histtype='stepfilled', alpha=0.2)
ax.set_xlim([x[0], x[-1]])
ax.legend(loc='best', frameon=False)
plt.show()
```
